```python
import os
styles = ["BAROQUE", "EARLY RENAISSANCE", "MEDIEVAL", "NEOCLASSICISM", "HIGH RENAISSANCE", "MINIMALISM", "REALISM",
          "IMPRESSIONISM", "ROCOCO", "SURREALISM", "MANNERISM", "ROMANTICISM",
         ]
styles_sum = {style: {"Train": 0, "Test": 0, "Validation": 0} for style in styles}

# Store data in dict
for style in styles:
    for dir in ['Train', 'Validation', 'Test']:
        styles_sum[style][dir] = len(os.listdir("../../sculpture_data/model_data/classes_12/{dir}/{style}".format(dir=dir, style=style)))
    styles_sum[style]['Total'] = sum([styles_sum[style][dir] for dir in styles_sum[style]])
        
```

```python
import pandas as pd
import numpy as np

styles_list = [{**{'style': style}, **styles_sum[style]} for style in styles_sum]
df = pd.DataFrame(styles_sum).T

df = df[['Total', 'Train', 'Test', 'Validation']]
df.index.name = "# of Images for style/set"
df = df.sort_values("Total", ascending=False)

df.head(12)
```
